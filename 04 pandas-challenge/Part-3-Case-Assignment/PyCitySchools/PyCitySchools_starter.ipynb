```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
df = pd.merge(student_data, school_data, how="left", on="school_name")

df.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
total_schools = len(school_data) 

total_students = len(student_data)

budget = school_data['budget'].sum()

avg_math = student_data['math_score'].mean()

avg_reading = student_data['reading_score'].mean()

x = student_data.loc[student_data['math_score'] >= 70]
x = x['math_score'].count()
pct_math = round(x / total_students * 100 , 2)

y = student_data.loc[student_data['reading_score'] >= 70]
y = y['reading_score'].count()
pct_reading = round(y / total_students * 100 , 2)

z = student_data.loc[(student_data['math_score'] >= 70) & (student_data['reading_score'] >= 70)]
z = z['student_name'].count()
overall_passing = round(z /total_students * 100 , 2)
```

```python
summary_df = pd.DataFrame([{'Total Schools': total_schools ,
                            'Total Students': total_students ,
                            'Total Budget': budget,
                            'Average Math Score': avg_math,
                            'Average Reading Score': avg_reading,
                            '% Passing Math': pct_math,
                            '% Passing Reading': pct_reading,
                            '% Overall Passing': overall_passing}])

summary_df['Total Students'] = summary_df['Total Students'].map('{:,}'.format)
summary_df['Total Budget'] = summary_df['Total Budget'].map('${:,.2f}'.format)
summary_df['Average Math Score'] = summary_df['Average Math Score'].map('{:.2f}'.format)
summary_df['Average Reading Score'] = summary_df['Average Reading Score'].map('{:.2f}'.format)
summary_df["% Passing Math"] = summary_df["% Passing Math"].map("{:.2f}%".format)
summary_df["% Passing Reading"] = summary_df["% Passing Reading"].map("{:.2f}%".format)
summary_df["% Overall Passing"] = summary_df["% Overall Passing"].map("{:.2f}%".format)

summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
groupby_school = df.groupby(['school_name'])

school_types = school_data.set_index('school_name')['type']

students_per = groupby_school['student_name'].count()

school_budget = school_data.set_index('school_name')
school_budget = school_budget['budget']

school_size = school_data.set_index('school_name')
school_size = school_size['size']

student_budget = school_budget/school_size

avg_byschool_math = round(groupby_school['math_score'].mean() , 2)

avg_byschool_reading = round(groupby_school['reading_score'].mean() , 2)

a = df.loc[(df['math_score'] >= 70)]
b = a.groupby('school_name')
c = b['student_name'].count()
pct_byschool_math = round(c/students_per * 100 , 2)

d = df.loc[(df['reading_score'] >= 70)]
e = d.groupby('school_name')
f = e['student_name'].count()
pct_byschool_reading = round(f/students_per * 100 , 2)

g = df.loc[(df['math_score'] >= 70) & (df['reading_score'] >= 70)]
h = g.groupby('school_name')
i = h['student_name'].count()
pct_byschool_overall = round(i/students_per * 100 ,2)
```

```python
school_summary_df = pd.DataFrame({'School Type': school_types,
                               'Total Students': students_per,
                               'Total School Budget': school_budget,
                               'Per Student Budget': student_budget,
                               'Average Math Score': avg_byschool_math,
                               'Average Reading Score': avg_byschool_reading,
                               '% Passing Math': pct_byschool_math,
                               '% Passing Reading': pct_byschool_reading,
                               '% Overall Passing': pct_byschool_overall})

school_summary_df['Total Students'] = school_summary_df['Total Students'].map('{:,}'.format)
school_summary_df['Total School Budget'] = school_summary_df['Total School Budget'].map('${:,.2f}'.format)
school_summary_df['Per Student Budget'] = school_summary_df['Per Student Budget'].map('${:,.2f}'.format)
school_summary_df['Average Math Score'] = school_summary_df['Average Math Score'].map('{:.2f}'.format)
school_summary_df['Average Reading Score'] = school_summary_df['Average Reading Score'].map('{:.2f}'.format)
school_summary_df['% Passing Math'] = school_summary_df["% Passing Math"].map("{:.2f}%".format)
school_summary_df['% Passing Reading'] = school_summary_df["% Passing Reading"].map("{:.2f}%".format)
school_summary_df['% Overall Passing'] = school_summary_df["% Overall Passing"].map("{:.2f}%".format)


school_summary_df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
bins = [ 0 , 584 , 629 , 644 , 675]
group_names = ['< $584' , '$585 - 629' , '$630 - 644' , '$645 - 675']

df['Budget per Student'] = df['budget']/df['size']

df['Spending Ranges (Per Student)'] = pd.cut(df['Budget per Student'], bins, labels=group_names)

groupby_spend = df.groupby('Spending Ranges (Per Student)')

math_avg = round(groupby_spend['math_score'].mean() , 2)
reading_avg = round(groupby_spend['reading_score'].mean() , 2)

n = df.loc[df['math_score'] >= 70]
n = n.groupby('Spending Ranges (Per Student)')
pct_math_pass = round(n['student_name'].count()/groupby_spend['student_name'].count() * 100 , 2)

o = df.loc[df['reading_score'] >= 70]
o = o.groupby('Spending Ranges (Per Student)')
pct_reading_pass = round(o['student_name'].count()/groupby_spend['student_name'].count() * 100 , 2)

p = df.loc[(df['reading_score'] >= 70) & (df['math_score'] >= 70)]
p = p.groupby('Spending Ranges (Per Student)')
pct_overall_pass = round(p['student_name'].count()/groupby_spend['student_name'].count() * 100 , 2)
```

```python
byschool_spending = pd.DataFrame({'Average Math Score' : math_avg,
                                  'Average Reading Score' : reading_avg,
                                  '% Passing Math' : pct_math_pass,
                                  '% Passing Reading' : pct_reading_pass,
                                  '% Overall Passing' : pct_overall_pass})

byschool_spending['% Passing Math'] = byschool_spending['% Passing Math'].map('{:,.2f}%'.format)
byschool_spending['% Passing Reading'] = byschool_spending['% Passing Reading'].map('{:,.2f}%'.format)
byschool_spending['% Overall Passing'] = byschool_spending['% Overall Passing'].map('{:,.2f}%'.format)

byschool_spending
```

```python
byschool_size = pd.DataFrame({'Average Math Score' : mathavg,
                                  'Average Reading Score' : readingavg,
                                  '% Passing Math' : pct_mathpass,
                                  '% Passing Reading' : pct_readingpass,
                                  '% Overall Passing' : pct_overallpass})

byschool_size['% Passing Math'] = byschool_size['% Passing Math'].map('{:,.2f}%'.format)
byschool_size['% Passing Reading'] = byschool_size['% Passing Reading'].map('{:,.2f}%'.format)
byschool_size['% Overall Passing'] = byschool_size['% Overall Passing'].map('{:,.2f}%'.format)

byschool_size
```
