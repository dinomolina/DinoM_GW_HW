### Python Mini-Project: Data Clean-Up, Pt. 2

```python
# Dependencies
import pandas as pd
```

```python
# load CSV
df = pd.read_csv("Resources/2016-FCC-New-Coders-Survey-Data.csv")
df.head()
```

```python
# Read with pandas--low_memory required to suppress errors about mixed data types
```

```python
# Take only columns 0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111

columns = [0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]
extract_df = df.iloc[ : , columns]
extract_df.head()
```

```python
# Change reading 0 and 1 to No and Yes, respectively

extract_df= extract_df.replace({0.0 : 'no', 1.0: 'yes'})

extract_df.head()
```

```python
# Extract rows for only those who attended a bootcamp

attendees_df = extract_df.loc[extract_df["AttendedBootcamp"] == "yes"]

attendees_df.head()
```

```python
# Create DataFrame of the different boot camps that had a significant number of attendees

bootcamp = attendees_df.groupby(["BootcampName"])

bootcamp_attendees = bootcamp["AttendedBootcamp"].count()

bootcampA_df= pd.DataFrame(bootcamp_attendees)

bootcampA_df = bootcamp_df.sort_values("AttendedBootcamp", ascending=False)

bootcampA_df = bootcampA_df.reset_index()
```

```python
# Count how many attendees of each bootcamp would recommend the bootcamp

recommend_df = attendees_df.loc[extract_df["BootcampRecommend"] == "yes"]

recommend = recommend_df.groupby(["BootcampName"])

bootcamp_recommend = recommend["BootcampRecommend"].count()

bootcampB_df= pd.DataFrame(bootcamp_recommend)

bootcampB_df = bootcampB_df.sort_values("BootcampRecommend", ascending=False)

bootcampB_df = bootcampB_df.reset_index()
```

```python
# Merge the two created data frames on the name of tbe bootcamp

merge_df = pd.merge(bootcampA_df, bootcampB_df, on = "BootcampName")

merge_df.head()
```

```python
# Calculate percentage of eac bootcamp's students who are recommenders

merge_df["% Recommend"] = round(merge_df['BootcampRecommend'] / merge_df['AttendedBootcamp'] *100 , 2)

merge_df = merge_df.sort_values("% Recommend", ascending=False)

merge_df = merge_df.rename(columns={"AttendedBootcamp":"Count", "BootcampRecommend":"Amount that Recommend"})
merge_df["% Recommend"] = merge_df["% Recommend"].map("{:.2f}%".format)

merge_df
```

```python
# Export to excel and remove index
merge_df.to_excel("Output/summary.xlsx", index=False, header=True)
```
