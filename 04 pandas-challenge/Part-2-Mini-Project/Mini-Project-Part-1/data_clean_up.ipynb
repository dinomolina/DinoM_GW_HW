```python
# Dependencies
import pandas as pd
```

```python
# load CSV
df = pd.read_csv("Resources/2016-FCC-New-Coders-Survey-Data.csv")
df.head()
```

```python
# Inspect all columns
df.count()
```

```python
# Extract only columns 0, 1, 2, 3, 4, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
columns = [0, 1, 2, 3, 4, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]
extract_df = df.iloc[ : , columns]
extract_df.head()
```

```python
#extract_df.AttendedBootcamp.replace(to_replace=[0 ,1], value=['no', 'yes'])

extract_df['AttendedBootcamp'] = extract_df['AttendedBootcamp'].replace({0.0 : 'no', 1.0: 'yes'})

extract_df.head()
```

```python
# Calculate total number of respondents in survey
total_respondents = len(extract_df['Age'])

total_respondents
```

```python
# Extract rows corresponding only to people who attended a bootcamp
attendees_df = extract_df.loc[extract_df["AttendedBootcamp"] == "yes"]

attendees = attendees_df["AttendedBootcamp"].count()

attendees
```

```python
# Calculate average age of attendees

avg_age = round(attendees_df['Age'].mean(),2)

avg_age
```

```python
# Calculate how many people attended a bootcamp

attendees = attendees_df['AttendedBootcamp'].count()

attendees
```

```python
# Calculate how many attendees hold degrees

attendees_df['SchoolDegree'] = attendees_df['SchoolDegree'].replace({"bachelor's degree" : 'degree',
                                                                 "some college credit, no degree" : 'no degree',
                                                                 "master's degree (non-professional)": 'degree',
                                                                 "professional degree (MBA, MD, JD, etc.)": 'degree',
                                                                 "high school diploma or equivalent (GED)": 'no degree',
                                                                 "associate's degree" : 'degree',
                                                                 "trade, technical, or vocational training": 'no degree',
                                                                 "some high school": 'no degree',
                                                                 "Ph.D.": 'degree',
                                                                 "no high school (secondary school)": 'no degree'})

degrees_df= attendees_df.loc[attendees_df["SchoolDegree"] == "degree", :]

attendees_degrees = degrees_df["SchoolDegree"].count()

attendees_degrees
```

```python
# Count number of attendees who self-identify as male; female; or are of non-binary gender identification

male_df = attendees_df.loc[extract_df["Gender"] == "male"]

male = male_df["Gender"].count()

female_df = attendees_df.loc[extract_df["Gender"] == "female"]

female = female_df["Gender"].count()

nonbinary_df = attendees_df.loc[(df["Gender"] == "genderqueer") | (df["Gender"] == "trans") | (df["Gender"] == "agender"), :]

nonbinary = nonbinary_df["Gender"].count()

total_gender = male + female + nonbinary

print(male)
print(female)
print(nonbinary)
print(total_gender)
```

```python
# Calculate percentage of respondents who attended a bootcamp

attendees_pct = round(attendees/total_respondents * 100 ,2)

attendees_pct
```

```python
# Calculate percentage of respondents belonging to each gender

male_pct = round(male/total_gender * 100 , 2)
female_pct = round(female/total_gender * 100 , 2)
nonbinary_pct = round(nonbinary/total_gender * 100 , 2)

print(male_pct)
print(female_pct)
print(nonbinary_pct)
```

```python
# Calculate percentage of attendees with a school degree
degree_pct = round(attendees_degrees /attendees *100 ,2)

degree_pct
```

```python
# Calculate average post-bootcamp salary
avg_salary = round(attendees_df['BootcampPostSalary'].mean() , 2)

avg_salary
```

```python
# Create a new table consolodating above calculations

summary_df =pd.DataFrame([{'Total Surveyed' : total_respondents,
                           'Total Bootcamp attendees' : attendees, 
                           '% attended bootcamp' : attendees_pct, 
                           'Avg. Age' : avg_age, 
                           'Has a degree' : degree_pct, 
                           '% Male' : male_pct, 
                           '% Female' : female_pct, 
                           '% Non Gender Specific' : nonbinary_pct, 
                           'Average Post Bootcamp Salary' : avg_salary}])

summary_df
```

```python
# Improve formatting before outputting spreadsheet

summary_df["Total Surveyed"] = summary_df["Total Surveyed"].map("{:,}".format)
summary_df["% attended bootcamp"] = summary_df["% attended bootcamp"].map("{:.2f}%".format)
summary_df["Has a degree"] = summary_df["Has a degree"].map("{:.2f}%".format)
summary_df["% Male"] = summary_df["% Male"].map("{:.2f}%".format)
summary_df["% Female"] = summary_df["% Female"].map("{:.2f}%".format)
summary_df["% Non Gender Specific"] = summary_df["% Non Gender Specific"].map("{:.2f}%".format)
summary_df["Average Post Bootcamp Salary"] = summary_df["Average Post Bootcamp Salary"].map("${:,.2f}".format)

summary_df
```
